```python
# Load data
import pickle
with open ('preprocessed_all_TR_50638.pickle', 'rb') as json_file:
    tweets = pickle.load(json_file)
```

```python
from gensim.summarization import keywords
```

```python
keys = []
for l in tweets:
  keys.append(keywords(l,split=True, lemmatize=True, deacc=True))
```

```python
count = 0
for k in keys:
  if k == []:
    count += 1
count
```

```python
print(keys[:10])
```

```python
import re
final = []
for k in keys:
  keys_one_word = []
  for i in k:
    ind = re.split('\s',i)
    keys_one_word.append(ind)
  final.append(keys_one_word)
```

```python
print(final[:10])
```

```python
z = []
for f in final:
  x = []
  for i in f:
    x = x+i
  z.append(x)
z
```

```python
len(z)
```

```python
# Create pickle
with open('tweets_keywords_50638.pickle', 'wb') as fp:
    pickle.dump(z, fp)
```
